```python
x = np.linspace(0, 1, 20)
y = 5 - 3 * x
plt.plot(x, y)
plt.title('y = 5 - 3x');
plt.xlabel('x')
plt.ylabel('y')
```

```python
x = np.linspace(-1, 1, 50)
y = x ** 2
# plt.scatter(x, y)
plt.plot(x, y)
plt.title('$y = x^{2}$')
plt.axhline(color = 'black',
            linestyle='--',
            linewidth=0.8)
plt.axvline(color='black',
            linestyle='--',
            linewidth=0.8);
```

```python
plt.rcParams['figure.figsize'] = [8, 8]
```

```python
plt.subplot(2, 2, 1)
x = np.linspace(1 / np.e ** 2, np.e ** 3)
y = np.log(x)
plt.plot(x, y)
plt.title(r'$y = \ln(x)$')


plt.subplot(2, 2, 2)
x = np.linspace(0, 3)
y = np.exp(x)
plt.plot(x, y)
plt.title(r'$y = e^{x}$')


plt.subplot(2, 2, 3)
x = np.linspace(0, 2 * np.pi)
y = np.sin(x)
plt.plot(x, y)
plt.title(r'$y = \sin(x)$')

plt.subplot(2, 2, 4)
x = np.linspace(1, 5)
y = 1 / x
plt.plot(x, y);
plt.title(r'$y = 1 / x$');
```
